#6. Filtrado de datos

1. Encuentra a los jugadores que cobran más de un millón de dólares. Usa una mask
2. Encuentra a los bases (position) de los Lakers.
3. Encuentra a los jugadores que estudiaron en Texas y juegan actualmente en Texas.


```python
#1
mask_salario = df_nba['Salary'] >=1000000
df_salario=df_nba[mask_salario]
df_salario
```

```python
#2
PG = (df_nba['Position'] == 'PG') & (df_nba['Team'] == 'Los Angeles Lakers')
df_nba[PG]
```

```python
#3
nba_texas = (df_nba ['College'].isin(['Texas', 'Texas A&M', 'Bylor', 'Houston'])) & ( (df_nba ['Team'] == 'Dallas Mavericks') | (df_nba ['Team'] == 'Houston Rockets') | (df_nba ['Team'] == 'San Antonio Spurs'))
df_nba[nba_texas]
```
