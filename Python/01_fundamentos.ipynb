### Ejercicio 1
Pide al usuario su nombre y edad con `input()` y guarda los valores en variables. Luego imprime un saludo como:
`Hola Ana, tienes 25 años.`
Usa una f-string para formatear la salida.

### Ejercicio 3
Pide al usuario que introduzca una temperatura en grados Celsius usando `input()`.
Convierte ese valor a tipo `float` y aplica la fórmula para convertirlo a Fahrenheit:

```python
F = C * 9/5 + 32
```

Muestra el resultado en pantalla indicando claramente ambos valores (Celsius y Fahrenheit) con una f-string.

````python
temperatura = float(input("Ingrese la temperatura: "))
Fahrenheit = temperatura * 9/5 + 32

print(f'{temperatura} Cº equivale a {Fahrenheit} Fº.')
````

### Ejercicio 4
Pide al usuario que introduzca un número con `input()` y convierte ese valor a `int`. Luego determina si el número es par o impar. Muestra el resultado con un mensaje claro usando `print()` y f-string si es posible.

````python
numero = int(input('Ingrese un numero: '))

print(f"{numero} es un numero {'par' if numero % 2 == 0 else 'impar'}.")
````
